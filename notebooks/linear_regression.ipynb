Import thư viện

```python
import numpy as np
import matplotlib.pyplot as plt
```

Tạo dữ liệu giả

```python
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

Khởi tạo tham số

```python
w = np.random.randn(1,1)
b = 0.0
lr = 0.05
epochs = 1000
```

Huấn luyện mô hình bằng gradient descent

B1: Đưa data vào hàm tuyến tính để dự đoán y
B2: Tính sai số giá trị thực tế và giá trị dự đoán (Dùng mean square error)
B3: Tính đạo hàm riêng (gradient) với tham số w và tham số b
B4: Cập nhật w và b mới bằng công thức W_t = W_t-1 - lr.dw
Lặp lại quá trình

```python
for epoch in range(epochs):
    y_pred = X.dot(w) + b
    error = y_pred - y
    loss = (error ** 2).mean()
    dw = (2 / len(X)) * X.T.dot(error)
    db = (2 / len(X)) * np.sum(error)
    
    w -= lr * dw
    b -= lr * db
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")
print("Final weights:", w)
print("Final bias:", b)
```

Vẽ kết quả

```python
plt.scatter(X, y, color = "blue", label='Data')
plt.plot(X, X.dot(w) + b, color = "red", label='Model')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression from Scratch")
plt.show()
```
